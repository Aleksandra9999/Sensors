```python
dX = [np.sum(acc[:i, 0]) + acc[i, 0]*(time[0]-time[1]) for i in range(len(acc))]
dY = [np.sum(acc[:i+1, 1]) + acc[i, 1]*(time[0]-time[1]) for i in range(len(acc))]

sns.lineplot(time, dX, label="dx")
sns.lineplot(time, dY, label="dy")
```
